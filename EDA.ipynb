```python
#feature engineering
```

```python
data["Date"]=data["Date_of_Journey"].str.split("/").str[0]
```

```python
data["Month"]=data["Date_of_Journey"].str.split("/").str[1]
```

```python
data["Year"]=data["Date_of_Journey"].str.split("/").str[2]
```

```python
data.head()
data["Year"]=data["Year"].astype(int)

data["Month"]=data["Month"].astype(int)

data["Date"]=data["Date"].astype(int)

data
```

```python
data.info()
```

```python
data["Arrival_Hour"]=data["Arrival_Time"].str.split(':').str[0]
data["Arrival_Min"]=data["Arrival_Time"].str.split(':').str[1]
```

```python
data
```

```python
data["Arrival_Min"]=data["Arrival_Min"].apply(lambda x:x.split()[0])
```

```python
data
```

```python
data["Arrival_Hour"]=data["Arrival_Hour"].astype(int)

data["Arrival_Min"]=data["Arrival_Min"].astype(int)
data.info()
```

```python
data["Dep_Hour"]=data["Dep_Time"].str.split(':').str[0]
data["Dep_Min"]=data["Dep_Time"].str.split(':').str[1]
```

```python
data.drop(columns="Route",inplace=True)
```

```python
data
```

```python
data["Dep_Hour"]=data["Dep_Hour"].astype(int)

data["Dep_Min"]=data["Dep_Min"].astype(int)
data.info()
```

```python
data["Total_Stops"]=data["Total_Stops"].map({'non-stop':0,'1 stop':1,'2 stop':2,'3 stop':3,'4 stop':4,np.nan:1})
```

```python
data["Total_Stops"]=data["Total_Stops"].fillna(1)
```

```python
data.info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder()
df=pd.DataFrame(encoder.fit_transform(data[["Airline","Source","Destination","Additional_Info"]]).toarray(),columns=encoder.get_feature_names_out())
```

```python
data=pd.concat([data,df])
data.drop(columns=["Airline","Source","Destination","Additional_Info"],inplace=True)
data.info()
```

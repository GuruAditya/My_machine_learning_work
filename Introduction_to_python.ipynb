```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a single subplot
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6, 4))

# Plot on the single axes
ax.plot(x, y, 'b-', label='sin(x)')
ax.set_title('Single Subplot')
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create two subplots (1 row, 2 columns)
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 4))
fig.set_size_inches(6,3)
# Plot on each axes (ax is a 1D array)
ax[0].plot(x, y1, 'b-.', label='sin(x)')
ax[0].set_title('Sine')
ax[0].set_xlabel('x')
ax[0].set_ylabel('sin(x)')
ax[0].legend()

ax[1].plot(x, y2, 'r-', label='cos(x)')
ax[1].set_title('Cosine')
ax[1].set_xlabel('x')
ax[1].set_ylabel('cos(x)')
ax[1].legend()

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create first subplot (top of 2x1 grid)
plt.subplot(2, 1, 1)
plt.plot(x, y1, 'b-')
plt.title('Sine')

# Create second subplot (bottom of 2x1 grid)
plt.subplot(2, 1, 2)
plt.plot(x, y2, 'r-')
plt.title('Cosine')

plt.tight_layout()
plt.show()
```
